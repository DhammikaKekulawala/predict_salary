```python
df = pd.read_csv("w2_position_salaries-220925-153437.csv")
df.head(5)
```

```python
df.shape
```

```python
X = df[['Level']]
y = df['Salary']
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X, y)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred = linear_reg.predict(X)
linear_r2 = r2_score(y, y_pred)

# Step 5: Print the R2 score for the Simple Linear Regression model
print("R2 Score - Simple Linear Regression:", linear_r2)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
r2_scores_polynomial = {}
degrees = [2, 3, 4]

for degree in degrees:
    poly_features = PolynomialFeatures(degree=degree)
    X_poly = poly_features.fit_transform(X)

    poly_reg = LinearRegression()
    poly_reg.fit(X_poly, y)

    # Compute the R2 score for the current polynomial regression model
    poly_r2 = poly_reg.score(X_poly, y)
    r2_scores_polynomial[f'Degree {degree}'] = poly_r2

# Step 6: Print the R2 scores
print("R2 Score - Simple Linear Regression:", linear_r2)
for degree, r2_score_poly in r2_scores_polynomial.items():
    print(f"R2 Score - Polynomial Regression (Degree {degree}):", r2_score_poly)
```
